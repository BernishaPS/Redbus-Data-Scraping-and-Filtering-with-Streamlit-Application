```python
driver_AP=webdriver.Chrome()
driver_AP.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")

time.sleep(2)

driver_AP.maximize_window()
```

```python
# Initialize WebDriverWait
wait = WebDriverWait(driver_AP, 20)

def andhra_link_route(path):
    LINKS_ANDHRA = []
    ROUTE_ANDHRA = []

    # Loop through the first 5 pages (or more if needed)
    for i in range(1,6):
        # Find elements by XPath
        paths = driver_AP.find_elements(By.XPATH, path)

        # Retrieve the route links
        for dlink in paths:
            d = dlink.get_attribute("href")
            LINKS_ANDHRA.append(d)

        # Retrieve the names of the routes
        for route in paths:
            ROUTE_ANDHRA.append(route.text)

        try:
            # Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))

            # Find the next button element for pagination
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()="{i+1}"]')
            time.sleep(3)  # Small wait for page transition
# Ensure the next button is clickable by using JavaScript (if .click() doesn't work)

            driver_AP.execute_script("arguments[0].scrollIntoView();", next_button) # Scroll to the button
            time.sleep(1) # Allow time for scrolling action
            
            # Try using JavaScript to click the button 
            driver_AP.execute_script("arguments[0].click();", next_button)
            #print(f"Clicked the next button for page [i+1] via JavaScript")
            time.sleep(3) # Optional sleep to wait for page transition

        

            #next_button.click()  # Click the next page button (-NOT WROKING-)

        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break

    return LINKS_ANDHRA, ROUTE_ANDHRA

# Example usage
LINKS_ANDHRA, ROUTE_ANDHRA = andhra_link_route("//a[@class='route']")
```

```python
df_AP=pd.DataFrame({"Route_name":ROUTE_ANDHRA,"Route_link":LINKS_ANDHRA})
```

```python
df_AP.to_csv('AP.csv',index = False)
```
